<h1 id="1."><b>1. Planteamiento del problema</b></h1>
<br>
<div align="justify">
Sea un electrón confinado en una película delgada, con una muy baja probabilidad de encontrarse fuera de la misma.  Si se asume una región de confinamiento en forma de paralepípedo regular, como una caja con lados $L_x$, $L_y$ y $L_z$, en la cual se tiene energía potencial eléctrica igual a cero dentro de la caja e infinita en los bordes y fuera de la misma, calcular la absorción óptica $\alpha(\omega)$ y el cambio en el índice de refracción $\Delta\eta(\omega)/\eta_r$  cuando una onda electromagnética plana monocromática con frecuencia angular $\omega$, campo eléctrico $\vec{F}$ y vector potencial $\vec{A}$ incide sobre la región donde está confinado el electrón.
<br><br>
En este documento todas las ecuaciones están en el sistema internacional de unidades, SI.
</div>

<h1 id="2."><b>2. Electrón en una caja de potencial</b></h1>
<br>
<div align="justify">
    Sea un electrón confinado en una película delgada de ancho $L_{z}$ y lados $L_{x}$, $L_{y}$, tales que $L_{z}\ll L_{x}$, $L_{z}\ll L_{y}$. Una onda electromagnética plana monocromática de frecuencia angular $\omega $ atraviesa la película delgada con una dirección de propagación paralela al eje $z$ y polarizada en una cierta dirección $\hat{u}_{p}$,

<center>$\hat{u}_{p}=\cos (\theta )\hat{i}+sen(\theta )\hat{j}\qquad$      (1)</center>

Como se supone que el electrón está en una película delgada, se puede asumir que la función de onda que describe su estado depende solo de las coordinadas que describen dicha película $(x,y)$.  Entonces, la función de onda $\Psi _{n_{x}n_{y}}(x,y)$ para un electrón en una 
caja de potencial de lados $L_{x}$ y $L_{y}$ está dada por,

<center>$\Psi _{n_{x}n_{y}}(x,y)=\varphi _{n_{x}}(x)\vartheta _{n_{y}}(y)\qquad$     (2)</center>

Donde,

<center>$\varphi _{n_{x}}(x)=\sqrt{\frac{2}{L_{x}}}sen\left(\frac{n_{x}\pi 
}{L_{x}}x\right)\qquad$   (3)</center>

<br>
<center>$\vartheta _{n_{y}}(y)=\sqrt{\frac{2}{L_{y}}}sen\left(\frac{n_{y}\pi}{L_{y}}y\right)\qquad$   (4)</center>


La energía está determinada por los dos números cuánticos $n_{x}$ y $n_{y}$,

<center>$E_{n_{x}n_{y}}=\frac{\pi^{2}\hbar^{2}}{2m} \left( \frac{n_{x}^{2}}{L_{x}^{2}}+\frac{n_{y}^{2}}{L_{y}^{2}} \right)\qquad$    (5)</center>
</div>

Para la implementación en Python, asumirá **phy_x** = $\varphi _{n_{x}}(x)$, **theta_y** = $\vartheta _{n_{y}}(y)$ y **Psi_xy** = $\Psi _{n_{x}n_{y}}(x,y)$

```python
import numpy as np
import pandas as pd
import scipy.constants as cte

Lx = 100 # nm
Ly = 100 # nm
Lz = 10  # nm

N = 50 # número de puntos

m = 0.8 * cte.m_e  # Masa efectiva del electrón en metros

delta_x = Lx/10.0
delta_y = Ly/10.0
x = np.linspace(-delta_x, Lx + delta_x, N)
y = np.linspace(-delta_x, Ly + delta_y, N)

xx, yy = np.meshgrid(x, y, sparse=True)

def phy_x(x, n_x):
    phy = np.sqrt(2.0/Lx)*np.sin(n_x*np.pi*x/Lx)
    phy[(x > Lx) | (x < 0)] = 0.0
    return phy

def theta_y(y, n_y):
    theta = np.sqrt(2.0/Ly)*np.sin(n_y*np.pi*y/Ly)
    theta[(y > Ly) | (y < 0)] = 0.0
    return theta

Psi_xy = lambda x, y, n_x, n_y: phy_x(x, n_x)*theta_y(y, n_y)
E = lambda n_x, n_y: (np.pi*cte.hbar)**2 * ((n_x/(Lx*1e-9))**2 + (n_y/(Ly*1e-9))**2) / (2.0*m)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

%matplotlib inline

fig = plt.figure()
ax = fig.gca(projection='3d')

nx = 1
ny = 2
ax.plot_surface(xx, yy, Psi_xy(xx, yy, nx, ny), cmap='viridis', edgecolor='none')
ax.axes.set_xlabel(r'$x$ $[nm]$')
ax.axes.set_ylabel(r'$y$ $[nm]$')
ax.axes.set_zlabel(r'$\Psi_{%d,%d}(x,y)$ $[nm]$' % (nx, ny))
ax.axes.set_zticks([])
print('E(%d,%d) = %0.3g [J] = %0.3g [eV]' % (nx, ny, E(nx,ny), E(nx,ny)/cte.e))
plt.show()
```

<h1 id="3."><b>3. Modelo semiclásico de la interacción de la luz con la materia</b></h1>
<br>
<div align="justify">
    Dado un cierto sistema cuántico descrito por el Hamiltoniano $\hat{H}_{0}$, con auto valores $E_{n}$ y funciones de estado $|n\rangle$, sobre el cual incide una onda electromagnética plana monocromática con frecuencia angular $\omega $, vector de onda $\vec{k}$, campo eléctrico $\vec{F}$ y vector potencial $\vec{A}=A\hat{u}_{p}$. Se supondrá que la onda electromagnética perturba al sistema descrito por $\hat{H}_{0}$ de tal manera que los observables pueden seguir siendo descritos por las funciones de estado $|n\rangle$. De esta manera, la interacción entre la onda electromagnática y el sistema cuántico se puede describir por el Hamiltoniano,

<br>

<center>$\hat{H}=\hat{H}_{0}+\frac{q_{e}}{m}\vec{A}\cdot \vec{p}\qquad$   (6)</center>


Donde $q_{e}$ es la carga del electrón, $m$, su masa y $\vec{p}=-j\hbar\vec{\nabla }$, el operador momentum.
</div>

<h2 id="3.1."><b>3.1. Probabilidad de transición por unidad de tiempo $R(\hbar\omega )$</b></h2>
<br>
<div align="justify">
Cuando la onda electromagnética incide sobre el sistema donde se encuentra confinado el electrón, este podrá absorber una cierta energía asociada con $\hat{H}_{1}=\frac{q_{e}}{m}\vec{A}\cdot \vec{p}$, la cual producirá que el electrón pase de un cierto estado $|i\rangle$ a otro $|f\rangle$. Las transiciones solo podrán ocurrir para las diferencias de energía permitidas, las cuales están dadas por $\Delta E_{if}=E_{f}-E_{i}$. Como existen varios estados probables para las transiciones, dados por la densidad de estados $g(E)=\frac{dN}{dE}$, donde $N$ es el número de partículas, será necesario considerar todas las transiciones probables para una cierta diferencia de energía $E_{if}$. La estimación de la probabilidad de transiciónpor unidad de tiempo asociada al operador $\hat{H}_{1}$ se puede describir por medio de la regla de oro de Fermi, 

<center>$R(\hbar\omega )=\frac{2\pi }{\hbar}\sum_{i}{\sum_{f > i}{\left|\left\langle 
f\right|\hat{H}_{1}\left|i\right\rangle \right|
^{2} g(E) \delta (\Delta E_{if}-\hbar\omega )}}$</center>


Donde $R(\hbar\omega )$ es la probabilidad de transición por unidad de tiempo y se asume que a pesar de estar en la aproximación semi-clásica, la energía de la onda electromagnética se absorbe en energías cuantizadas $\hbar\omega$, es decir, fotones; en otras palabras, se absorve energía de la onda electromagnética cuando se cumple que $\Delta E_{if}=\hbar\omega $. Note que $g(E)$, para ciertos casos, puede ser aproximada por una distribución uniforme, donde $g_0$, con undad de inverso de energía, corresponde con un valor constante (esto es una aproximación) para todos los estados a considerar,

<center>$R(\hbar\omega )=\frac{2\pi }{ \hbar}\sum_{i}{\sum_{f > i}{\left|\left\langle 
f\right|\hat{H}_{1}\left|i\right\rangle \right|^{2} g(E) \delta (\Delta E_{if}-\hbar\omega )}}$</center>


Es conveniente reescribir la expresión $\sum_{i}{\sum_{f > i}{\left|\left\langle f\right|\hat{H}_{1}\left|i\right\rangle \right|^{2} g(E) \delta (\Delta E_{if}-\hbar\omega )}}$, para ello se debe considerar que $\hat{H}_{1}=\frac{q_{e}}{m}\vec{A}\cdot \vec{p}=-j\hbar A\frac{q_{e}}{m}(\hat{u}_{p}\cdot \vec{\nabla })$. Si se define $\rho (\omega )$ como, 

<center>$\rho (\omega )=\sum_{i}{\sum_{f > i}{\left|\left\langle 
f\right|\hat{u}_{p}\cdot \vec{\nabla }\left|i\right\rangle \right|^{2} g(E) \delta (\Delta E_{if}-\hbar\omega 
)}}\qquad$ (7)</center>

<br>
entonces,

<center>$R(\hbar\omega )=2\pi\hbar A^{2}\left(\frac{q_{e}}{m}\right)^{2}\rho 
(\omega )\qquad$ (8)</center>
  
Note que las unidades de $R(\hbar \omega)$ son de inverso del tiempo $(s^{-1})$.
<br>
</div>

<h2 id="3.2."><b>3.2. Pérdida de potencia por unidad de volumen</b></h2>
<br>
<div align="justify">
La pérdida de potencia de una onda electromagnética que se desplaza a una velocidad $v$ y atraviesa un cierto volumen $\mho$ en un intervalo de tiempo $\Delta t$ está dada por la variación de su intensidad $-\Delta I \frac{\mho}{v\Delta t} = -\frac{dI}{dt}\Delta t\frac{\mho}{v\Delta t}$; donde $\frac{\mho}{v\Delta t}$ es el área del frente de onda con intensidad $I$. Asumiendo que se puede aplicar el principio de igual probabilidad a priori, entonces, todas las transiciones posibles tendrán la misma probabilidad, por lo tanto, la pérdida de energía de la onda se puede aproximar con $\hbar\omega $ veces la probabilidad de transición por unidad de tiempo $R(\hbar\omega)$, es decir,

<center>$-\frac{dI}{dt}\cong \frac{v}{\mho}\hbar\omega R = \frac{v}{\mho}\hbar\omega R \qquad$   (9)</center>

Para calcular la intensidad $I$ es necesario estudiar la propagación de las ondas electromagnéticas en medios con pérdidas.
</div>

<h2 id="3.3.">3.3. Ondas electromagnéticas</h2>
<div align="justify">
<br>
    Una onda electromagnética plana monocromática está descrita por su frecuencia angular $\omega $, vector de onda $\overrightarrow{k}$, campo eléctrico $\overrightarrow{F}$, vector potencial $\overrightarrow{A}$ y campo magnético $\overrightarrow{B}$. Una onda electromagnética puede propagarse en el vacío, en ausencia de medio material y cuando se propaga en un medio material, los parámetros que la describen se ven afectados por dicho medio.

Una onda electromagnética monocromática que ingresa en un medio material puede ser absorbida o dispersada; en el primer caso, absorción, cambia la magnitud de los campos que definen su función de onda, en el segundo, dispersión, además de la magnitud cambia la direccián de propagacián y, para el caso de interacciones inelásticas, su frecuencia.

Las ecuaciones de Maxwell en ausencia de cargas externas permiten describir los campos eláctricos y magnéticos de una onda electromagnética. A continuación, se describe la forma explícita de los campos eléctrico, magnético y los potenciales vectores y escalar asociados a una onda electromagnética, además de la consecuencia de la interacción de dicha onda con un medio con pérdidas.
</div>

<h3 id="3.3.1">3.3.1. Ecuaciones de Maxwell en un medio con pérdidas</h3>
<div align="justify">
<br>
Dado un cierto medio con carga neta igual a cero, las ecuaciones de 
Maxwell, en SI, están dadas por:

<center>$\vec{\nabla }\cdot \vec{F}=0$</center>


<center>$\vec{\nabla }\cdot \vec{B}=0$</center>


<center>$\vec{\nabla }\times \vec{F}=-\frac{\partial \vec{B}}{\partial t}$</center>


<center>$\vec{\nabla }\times \vec{B}=\mu \vec{J}+\mu \varepsilon _{r}\frac{\partial \vec{F}}{\partial t}$</center>
</div>

<h3 id="3.3.2">3.3.2. Onda plana</h3>
<div align="justify">
<br>
Una onda plana se caracteriza porque el parámetro oscilante asociado a ella es homogéneo en una la región definida por un plano. Para el caso de la descripción del potencial vector de una onda electromagnética plana, este sería de la forma,

<center>$\vec{A}=A\hat{u}_{p}e^{j(\vec{q}\cdot \vec{r} \pm \omega t)}\qquad$        (10)</center>


Donde A es la magnitud del potencial vector en unidades de $Vs/m$, $j=\sqrt{-1}$, el número imaginario, $\omega $, la frecuencia angular, $\vec{k}$, el vector de onda, $\vec{r}$, vector posición, $t$, el tiempo y ${u}_{p}$, un vector unitario que define la polarización de la onda. Además, el signo $"\pm "$ se utiliza para recordar que la onda electromagnética se puede propagar en los dos posibles sentidos definida una cierta dirección. Las expresiones para los campos eléctricos y magnéticos son:

<center>$\vec{F}=F_{0}\vec{u}_{p}e^{j(\vec{k}\cdot \vec{r} \pm \omega t)}$</center>


<center>$\vec{B}=B_{0}\vec{u}_{p'}e^{j(\vec{k}\cdot \vec{r} \pm \omega t)}$</center>
</div>

<h3 id="3.3.3">3.3.3. Permitividad compleja $\varepsilon^{*}$</h3>
<div align="justify">
<br>
Si la onda electromagnética se propaga por un medio con pérdidas dondeexiste una densidad de corriente $\vec{J}$, se puedeaproximar, para el caso de medio homogéneo, que la densidad de corrientees proporcional al campo eléctrico, ley de Ohm, donde la conductividad $\sigma $ es la constante de proporcionalidad,

<center>$\vec{J}=\sigma \vec{F}$</center>


En este caso, el rotacional del campo eléctrico para una ondaelectromagnética plana se puede escribir como,

<center>$\vec{\nabla }\times\vec{B}=\vec{k}\times \vec{B}=\mu(\sigma +j\omega \varepsilon _{r})\vec{F}$</center>


Se define la conductividad compleja $\sigma ^{*}$ como,

<center>$\sigma ^{*}=\sigma +j\omega \varepsilon _{r}\qquad (11)$</center>


La conductividad compleja $\sigma ^{*}$ se relaciona con lapermitividad compleja $\varepsilon ^{*}$,

<center>$\sigma ^{*}=j\omega \varepsilon ^{*}\qquad (12)$</center>


<center>$\varepsilon ^{*}=\varepsilon_{r}-j\frac{\sigma }{\omega }\qquad (13)$</center>


En términos de la permitividad compleja,

<center>$\vec{k}\times \vec{B}=j\omega \mu \varepsilon^{*}\vec{F}$</center>

</div>

<h3 id="3.3.4">3.3.4. Índice de refracción $\eta^{*}$</h3>
<div align="justify">
<br>
La velocidad con la que se propaga una onda electromagnética dentro de un medio está determinada por la raíz cuadrada del inverso del producto entre la permitividad compleja $\varepsilon^{*}$ y la permeabilidad magnética $\mu$,

<center>$v=\left| \frac{1}{\sqrt{\mu \varepsilon ^{*}}}\right| $</center>


Dado que la velocidad de la luz está dada por $\frac{1}{\sqrt{\mu_{0}\varepsilon _{0}}}$, entonces, el índice de refracción complejo está dado por,

<center>$\eta ^{*}=\sqrt{\frac{\mu \varepsilon ^{*}}{\mu _{0}\varepsilon_{0}}}=c\sqrt{\mu \varepsilon ^{*}}$</center>


Para ondas electromagnéticas que se propagan en medios, el índice de refracción es una cantidad compleja $\eta ^{*}$ cuya parte real $\eta_{r}$, comúnmente conocido como índice de refracción, es el cociente entre la velocidad de la luz y la velocidad de la onda electromagnética en el medio y la parte imaginaria $\kappa $ corresponde con el coeficiente de extinción,

<center>$\eta ^{*}=\eta _{r}+j\kappa\qquad (14)$</center>


El cuadrado del índice de refracción complejo $\eta ^{*}$ está
relacionado con la permitividad compleja,

<center>$(\eta ^{*}) ^{2}=c^{2} \mu \varepsilon ^{*}$</center>


Reemplazando las componentes de estas cantidades complejas,


<center>$\eta _{r}^{2}-\kappa ^{2}+2j\eta _{r}\kappa =\mu c^{2} (\varepsilon_{r}+j\varepsilon _{I})$</center>


Se obtiene que la parte imaginaria de la permitividad $\varepsilon _{I}$ es,

<center>$\varepsilon _{I}=\frac{2}{\mu c^{2}}\eta _{r}\kappa $</center>


La parte real de la permitividad eléctrica está dada por,

<center>$\varepsilon _{r}=\frac{\eta _{r}^{2}-\kappa ^{2}}{\mu c^{2}}$</center>


Si se define $\varepsilon _{1}$, $\varepsilon _{2}$ tal que cumpla la siguiente relación y se asume que el medio no es magnético de tal manera que $\mu =\mu _{0}$ y $c^{2}=1/\mu _{0}\varepsilon _{0}$,

<center>$\varepsilon _{0}(\varepsilon _{1}+j\varepsilon _{2})=\varepsilon_{r}+j\varepsilon _{I}$</center>


Entonces,

<center>$\varepsilon _{1}=\eta _{r}^{2}-\kappa ^{2}\qquad (15)$</center>


<center>$\varepsilon _{2}=2\eta _{r}\kappa\qquad (16)$</center>
</div>

<h3 id="3.3.5">3.3.5. Relaciones entre $\vec{A}$, $\vec{B}$ y $\vec{F}$</h3>
<div align="justify">
<br>
A partir de las leyes de Faraday $\vec{\nabla }\times \vec{F}=-\frac{\partial \vec{B}}{\partial t}$ y Ampere-Maxwell $\vec{\nabla }\times\vec{B}=\frac{\eta ^{2}}{c^{2}}\frac{\partial\vec{E}}{\partial t}$ en un medio con índice de refracción $\eta$, se obtienen las relaciones entre el campo eléctrico $\vec{F}$ y el magnético $\vec{B}$,

<center>$\vec{k}\times \vec{F}=j\omega \vec{B}$</center>


<center>$\vec{B}\times \vec{k}=j\omega \mu \varepsilon^{*}\vec{F}$</center>


Nótese que las anteriores ecuaciones implican que,

<center>$k=\omega \sqrt{\mu \varepsilon^{*}}$</center>


Además, dado que $\vec{B}=\nabla \times\vec{A}$, se puede escribir el potencial vector $\vec{A}$ en términos del campo eléctrico $\vec{F}$ y magnético $\vec{B}$ como,

<center>$\vec{k}\times \vec{A}=\vec{B}\qquad (17)$</center>


<center>$\vec{A}=\frac{1}{j\omega }\vec{F}\qquad (18)$</center>


En términos del índice de refracción complejo $\eta ^{*}$:

<center>$\vert \vec{k}\vert =k=\frac{\omega }{c}\eta^{*}=\frac{\omega }{c}\eta _{r}+j\frac{\omega }{c}\kappa\qquad (19)$</center>
</div>

<h3 id="3.3.6.">3.3.6. Vector de Poynting $\vec{S}$ e intensidad $I$ de una onda plana</h3>
<div align="justify">
<br>
El vector de Poynting $\vec{S}$ representa el flujo direccional de energía de la onda electromagnética y está dado por,

<center>$\vec{S}=\frac{1}{\mu }\vec{F}\times\vec{B}$</center>


Para una onda electromagnética plana que se tiene,

<center>$\vec{S}=\frac{\omega }{\mu }A^{2}\vec{k}$</center>


El valor medio del vector de Poynting en un periodo corresponde con la
intensidad $I$ de la onda electromagnética,

<center>$I=\frac{\omega k}{2\mu }A^{2}=\frac{\omega ^{2}\eta }{2\mu c}A^{2}$</center>

</div>

<h3 id="3.3.7.">3.3.7. Coeficiente de absorción $\alpha$</h3>
<div align="justify">
<br>
Al escribir el potencial vector de una onda electromagnética en un medio con pérdidas se tiene,

<center>$\vec{A}=(A_{0}e^{-\frac{\omega }{c}\kappa {u}_{q}\cdot\vec{r}}){u}_{p}e^{j(\frac{\omega }{c}{u}_{q}\cdot\vec{r}\pm \omega t)}$</center>


Si la onda electromagnética se propaga en dirección paralela al eje $x$,

<center>$\vec{A}=(A_{0}e^{-\frac{\omega }{c}\kappa x}){u}_{p}e^{j(\frac{\omega }{c}x \pm \omega t)}$</center>


Note que la amplitud del potencial vector $\vec{A}$ de la onda electromagnética es atenuada en un factor $e^{- \frac{\omega}{c}\kappa x}$. La intensidad de la onda electromagnética antes de ingresar al medio $I_{0}$ y a una distancia $x$ de la frontera del medio, $I$, se relacionan por la ley de Lambert-Beer,

<center>$I=I_{0}e^{-\alpha x}\qquad (20)$</center>

Donde la intensidad $I$ corresponde con el valor medio del vector de Poynting $\vec{S}=\frac{1}{\mu }\vec{F}\times \vec{B}$ en un periodo, el cual se puede calcular considerando las ecuaciones (17) a (19),

<center>$I=\frac{\omega k}{2\mu }\vert A\vert ^{2}=\frac{\omega ^{2}\vert \eta^{*}\vert }{2\mu c}\vert A\vert ^{2}\qquad (21)$</center>


Como $I_{0}=\frac{\omega ^{2}\vert \eta ^{*}\vert }{2c\mu }\vert A_{0}\vert ^{2}$ y $\vert A\vert ^{2}=\vert A_{0}\vert^{2}e^{-2\frac{\omega }{c}\kappa x}$, entonces, al comparar con la ley de Lambert-Beer,

<center>$\frac{\omega ^{2}\vert \eta ^{*}\vert }{2c\mu }\vert A_{0}\vert^{2}e^{-2\frac{\omega }{c}\kappa x}=\frac{\omega ^{2}\vert \eta ^{*}\vert}{2c\mu }\vert A_{0}\vert ^{2}e^{-\alpha x}$</center>


Entonces, el coeficiente de absorción se relaciona con el coeficiente de extinción,

<center>$\alpha =2\frac{\omega }{c}\kappa\qquad (22)$</center>
</div>

<h3 id="3.3.8.">3.3.8. Cambio en el índice de refracción $\Delta\eta/\eta_r$</h3>
<div align="justify">
<br>
El cambio en el índice de refracción $\Delta \eta /\eta _{r}$, se define como,

<center>$\frac{\Delta \eta }{\eta _{r}}=\frac{\varepsilon _{1}-1}{2\eta _{r}}\qquad  (23)$</center>
</div>

<h1 id="4."><b>4. Coeficiente de absorción $\alpha(\omega)$ en la aproximación semi-clásica</b></h1>
<br>
<div align="justify">
Para calcular el coeficiente de absorción en la aproximación semi-clásica basta con analizar la relación entre la intensidad de la onda electromagnética con la pérdida de energía con undad de volumen, ecuación (9).  En tal sentido, la variación en el tiempo de la intensidad de la onda electromagnética a lo largo del camino óptico se puede calcular por medio de la regla de la cadena $\frac{dI}{dt}=\frac{dI}{dx}v$, donde $v$ es la velocidad de la onda electromagnética en el medio, dada por $v=\frac{c}{\vert \eta^{*}\vert }$. A partir de las ecuaciones (21), (8), (9), la ley de Lambert-Beer en forma diferencial $(\frac{dI}{dx}=-\alpha I)$ , y teniendo en cuenta que $\frac{dI}{dt}=\frac{c}{\vert \eta ^{*}\vert}\frac{dI}{dx}$, se obtiene que el coeficiente de absorción puede ser expresado como $\alpha =\frac{2\mu C}{\omega \vert \eta^{*}\vert \mho A^{2}} \hbar R$,

<center>$\alpha =4\pi\hbar^2 \frac{\mu c}{\omega \vert \eta^{*}\vert \mho} 
\left(\frac{q_e}{m}\right)^{2}\rho (\omega )\qquad$   (24)</center>
</div>

<h2 id="4.1."><b>4.1. Elementos matriciales</b></h2>
<br>
<div align="justify">
Para calcular el coeficiente de absorción $\alpha $, es necesario
determinar los elementos matriciales $\gamma _{if}=\left\langle f\right|\hat{u}_{p}\cdot
\vec{\nabla}\left|i\right\rangle$ con los cuales se define $\rho
(\omega )$, ver ecuaciones (7) y (24). De esta manera, los elementos
matriciales $\gamma _{if}$ están dados por,
    

<center>$\rho (\omega )=\sum_{i}{\sum_{f > i}{\left|\gamma _{if}\right|^{2}g(E)\delta (\Delta E_{if}-\hbar\omega 
)}}\qquad$</center>

<br>
donde, 
<center>$\gamma _{if}=\left\langle f\right|\hat{u}_{p}\cdot
\vec{\nabla}\left|i\right\rangle = 
\frac{4}{L_{x}L_{y}}\int_{0}^{L_{x}}{\int_{0}^{L_{y}}{sen\left(\frac{n'_{y}\pi}{L_{y}}y\right)}
sen\left(\frac{n'_{x}\pi }{L_{x}}x\right)
\left\lbrack 
cos\left(\theta\right)
\frac{\partial }{\partial x} +
sen\left(\theta \right)
\frac{\partial }{\partial y}
\right\rbrack 
sen\left(\frac{n_{y}\pi }{L_{y}}y\right)
sen\left(\frac{n_{x}\pi}{L_{x}}x\right)dxdy}$</center>

<br>
Al realizar las derivadas parciales se obtiene,
<center>$\gamma _{if}=\left\langle n_{x}^{'}n'_{y}\right|\hat{u}_{p}\cdot
\vec{\nabla}\left|n_{x}n_{y}\right\rangle = 
\frac{4 cos(\theta)}{L_{x}L_{y}}
\frac{n_{x}\pi }{L_{x}}
\int_{0}^{L_{x}}{
  sen\left(\frac{n'_{x}\pi }{L_{x}}x\right)
  cos\left(\frac{n_{x}\pi}{L_{x}}x\right)dx}
\int_{0}^{L_{y}}{
  sen\left(\frac{n'_{y}\pi}{L_{y}}y\right)
  sen\left(\frac{n_{y}\pi }{L_{y}}y\right)dy} + 
\frac{4 sen(\theta)}{L_{x}L_{y}}
\frac{n_{y}\pi}{L_{y}}
\int_{0}^{L_{y}}{
  sen\left(\frac{n'_{y}\pi }{L_{y}}y\right)
  cos\left(\frac{n_{y}\pi }{L_{y}}y\right)dy}
\int_{0}^{L_{x}}{
  sen\left(\frac{n'_{x}\pi}{L_{x}}x\right)
  sen\left(\frac{n_{x}\pi }{L_{x}}x\right)dx}$</center>


Nótese que hay dos tipos de integrales, las cuales corresponden a,

<center>$\int_{0}^{L}{
  sen\left(\frac{n'\pi }{L}\chi \right)
  sen\left(\frac{n\pi }{L}\chi \right) d\chi
}=\frac{L}{2}\delta _{n^{'}, n} $</center>

<br>

<center>$\int_{0}^{L}{
  sen\left(\frac{n'\pi }{L}\chi \right)
  cos\left(\frac{n\pi }{L}\chi \right)d\chi }=
  \frac{L}{\pi }
    \left[
        \frac{I_{1}(n^{'},n)}{n^{'}+n}+\frac{I_{2}(n^{'},n)}{n^{'}-n}
    \right]$</center>


Donde $n^{'}\ne n$, además,

<center>$I_{1}(n^{'},n) =\bigg\{
\begin{gathered}
1 \quad\mbox{para } (n^{'}+n) \mbox{ impar} \\
0 \quad\mbox{para otros valores} \\
\end{gathered}
$</center>


<center>$I_{2}(n^{'},n)=\bigg\{
\begin{gathered}
1 \quad\mbox{para } (n^{'}-n) \mbox{ impar} \\
0 \quad\mbox{para otros valores} \\
\end{gathered}
$</center>

<br>
<center>$\gamma _{if}=
  \left\langle n_{x}^{'}n^{'}_{y}\right|\hat{u}_{p}\cdot
\vec{\nabla}\left|n_{x}n_{y}\right\rangle
  = 2 cos\left(\theta\right)
  \frac{n_{x}}{L_{x}}
  \left[
      \frac{I_{1}(n_{x}^{'},n_{x})}
         {n_{x}^{'}+n_{x}}
     +\frac{I_{2}(n_{x}^{'},n_{x})}
         {n_{x}^{'}-n_{x}}
  \right]\delta
_{n_{y}^{'}, n_{y}} +
    2 sen\left(\theta\right)\frac{n_{y}}{L_{y}}
    \left[
        \frac{I_{1}(n_{y}^{'},n_{y})}
            {n_{y}^{'}+n_{y}}
       +\frac{I_{2}(n_{y}^{'},n_{y})}
             {n_{y}^{'}-n_{y}}
     \right]\delta
_{n_{x}^{'}, n_{x}}$</center>

De esta manera, $\rho$ está dado por,

<center>$\rho (\omega )=
  \sum_{n_x, n_y}{\sum_{(n_{x}^{'}, n_{y}^{'}) > (n_{x}, n_{y})}{
       \left|
             \left\langle n_{x}^{'}n^{'}_{y}\right|\hat{u}_{p}\cdot
\vec{\nabla}\left|n_{x}n_{y}\right\rangle
       \right|^{2}
       g(E) \delta \left(E_{n_{x}^{'}, n_{y}^{'}} - E_{n_{x}, n_{y}} -\hbar\omega 
\right)}}\qquad$</center>


</div>

<h2 id="4.2."><b>4.2. Reglas de selección</b></h2>
<br>
<div align="justify">
Para calcular el coeficiente de absorción $\alpha$ es necesario determinar los valores diferentes de cero en la sumatoria que define $\rho(\omega)$.  Los valores permitidos para $\omega$ son aquellos donde $\Delta E = \hbar\omega$, donde,

<center>$\Delta E = E_{n_{x}^{'}, n_{y}^{'}} - E_{n_{x}, n_{y}} =\frac{\pi^{2}\hbar^{2}}{2m} \left( \frac{{n_{x}^{'}}^2 - n_{x}^{2}}{L_{x}^{2}}+\frac{{n_{y}^{'}}^2 - n_{y}^{2}}{L_{y}^{2}} \right)\qquad$</center>


Los términos diferentes de cero en la sumatoria cumplen con las siguientes propiedades:

<ul style="list-style-type:disc;">
  <li>Solo son diferente de cero los números cuánticos $(n'+n)$ impares y los $(n'-n)$ impares.</li>
  <li>Contribución de la componente en $x$ de la onda electromagnética, $cos(\theta)$:</li>
  <ul style="list-style-type:square;">
      <li>Siempre el estado final $n_x^{'}$ debe ser mayor que el inicial $n_x$.</li>
      <li>Solo contribuyen los estados $n_{y}^{'} = n_y$.</li>
  </ul>
  <li>Contribución de la componente en $y$ de la onda electromagnética, $sen(\theta)$:</li>
  <ul style="list-style-type:square;">
      <li>Siempre el estado final $n_y^{'}$ debe ser mayor que el inicial $n_y$.</li>
      <li>Solo contribuyen los estados $n_{x}^{'} = n_x$.</li>
  </ul>
  <li>Se asumirá que hay un solo electrón disponible para la transición, por lo tanto, solo se considerará la primera transición para cada </li>
</ul>  
</div>

<h2 id="4.3."><b>4.3. Onda polarizada en dirección $x$: $\theta=0$</b></h2>
<br>
<div align="justify">
Al tomar $\theta = 0$ se obtiene:
    
<center>$\Delta E_{{n_{x}^{'}}, n_{x}} =\frac{\pi^{2}\hbar^{2}}{2m} \left( \frac{{n_{x}^{'}}^2 - n_{x}^{2}}{L_{x}^{2}}\right)\qquad$</center>

<br>

En el caso de $\rho$,
<center>$\rho (\omega )=
  \sum_{n_x}{\sum_{n_{x}^{'} > n_{x}}{
       4
      \left(\frac{n_{x}}{L_{x}}\right)^2
      \left[\frac{I_{1}(n_{x}^{'},n_{x})}{n_{x}^{'}+n_{x}}+\frac{I_{2}(n_{x}^{'},n_{x})}{n_{x}^{'}-n_{x}}\right]^2
       g(E) \delta \left(\Delta E_{{n_{x}^{'}}, n_{x}} -\hbar\omega 
\right)}}\qquad$</center>

<br>

Dado que solo los valores impares son permitidos, se pueden definir unas nuevas variables $l_1$ y $l_2$ que solo toman valores impares,

<center>$n_{x}^{'}+n_{x} = 2l_1 + 1$</center>
<center>$n_{x}^{'}-n_{x} = 2l_2 + 1$</center>

Entonces, 

<center>$\Delta E_{l_2, l_1} =\frac{\pi^{2}\hbar^{2}}{2mL_{x}^{2}} (2l_1+1)(2l_2+1)\qquad$</center>


Además, si se tiene en cuenta que cada transición se observa como un pico con un cierto ancho $\Gamma$, se puede aproximar $\delta\left(\Delta E_{{n_{x}^{'}}, n_{x}} -\hbar\omega 
\right)$ con una distribución de Cauchy, también conocida como lorentciana o distribución de Cauchy-Lorentz,

<center>$\mathscr{L}_{l_1,l_2}(\omega; \Gamma) = 
        \frac{1}{\pi}
        \frac{\frac{\Gamma}{2}}
         {\left(
             \hbar\omega-\Delta E_{l_2, l_1}
         \right)^2 
         + \left(\frac{1}{2}\Gamma\right)^2}$</center>



De esta manera, $\rho$ se puede reescribir en términos de las nuevas variables así, donde se asume que $g(E) = g_0$ es una constante,

<center>$\rho (\omega )=
  \sum_{l_1}{\sum_{l_2 > 0}^{l_1-1}{
       4
      \left(\frac{l_1-l_2}{L_{x}}\right)^2
      \left[\frac{1}{2l_1 + 1}+\frac{1}{2l_2 + 1}\right]^2
       g_0 \mathscr{L}_{l_1,l_2}(\omega; \Gamma)}}\qquad$</center>

El coeficiente de absorción estará determinado por,
  <center>$\alpha = \frac{4\pi\hbar^2\mu c}{\omega \vert \eta^{*}\vert L_xL_yL_z} 
\left(\frac{q_e}{m}\right)^{2}\rho (\omega )$</center>
  

Para la implementación en iPython se usará:
<ul style="list-style-type:disc;">
    <li>w para representar a $\omega$.</li>
    <li>l1 para $l_1$.</li>
    <li>l2 para $l_2$.</li>
    <li>Delta_E  para $\Delta E_{l_2, l_1}$.</li>
    <li>Gamma  para $\Gamma$.</li>
    <li>LL para $\mathscr{L}_{l_1,l_2}(\omega; \Gamma)$.</li>
    <li>rho para $\rho$.</li>
    <li>a para $\alpha$.</li>
</ul>
</div>

```python
try:
    pi = np.pi
except:
    import numpy as np
try:
    hbar = cte.hbar
except:
    import scipy.constants as cte
try:
    h = plt.figure()
    plt.close('all')
except:
    from mpl_toolkits.mplot3d import Axes3D
    import matplotlib.pyplot as plt

Gamma = cte.e * 50e-3 # 50 meV en Joules
eta = 1.1 # Índice de refracción
m = 0.8 * cte.m_e  # Masa efectiva del electrón en metros

E_min = 0.1 # eV
E_max = 0.5   # eV

Lx_min = 5 # nm
Lx_max = 10 # nm

numSimulations = 10
Lx_values = np.linspace(Lx_min, Lx_max, numSimulations)

Lx = 7 # nm -> Lx_values[0]
Ly = 100 # nm
Lz = 10  # nm

g0 = 1.0# Densidad de Estados, aproximación de densidad Uniforme
```

```python
w = np.linspace(E_min, E_max, 1000)*cte.e / cte.hbar # Frecuencias para energías entre 0 y 2 eV
E_photon = cte.hbar*w/cte.e # eV
l1 = np.arange(2,100)
l2 = lambda l1: l1 - 1 # Una sola transición posible, al estado más cercano a l_1
Delta_E  = lambda l1, l2: (np.pi*cte.hbar)**2 * (2*l2+1)*(2*l1+1)/(2*m*(Lx*1e-9)**2)
LL = lambda w, l1, l2: (0.5*Gamma/np.pi) / ((cte.hbar*w - Delta_E(l1,l2))**2 + (0.5*Gamma)**2)

plt.figure(dpi=200)
for lf in l1[:4]:
    plt.plot(E_photon, LL(w, lf, l2(lf)), 
             label=r'$\Delta E_{%d, %d} = %0.3f eV$' % (lf, l2(lf), Delta_E(lf, l2(lf))/cte.e))
    
plt.title(r'$L_x = %0.1f nm, $L_y = %0.1f nm, $L_z = %0.1f nm$, ' % (Lx, Ly, Lz))
plt.xlabel(r'$\hbar \omega$ $[eV]$')
plt.ylabel(r'$\mathscr{L}_{l_1,l_2}$')
plt.legend()
plt.show()
    
```

```python
def absorption(Lx):
    rho = 0*w
    for lf in l1:
        l1_l2 = lf-l2(lf)
        rho_f1 = 4*(l1_l2/(Lx*1e-9))**2
        rho_f1 *= (1.0/(2.0*lf + 1.0) + 1.0/(2.0*l2(lf)+1.0))**2
        rho_f1 *= g0 * LL(w, lf, l2(lf))
        rho += rho_f1
    hh = 4.0 * np.pi *cte.mu_0 * cte.hbar**2 * cte.c
    hh /= (w*eta)*(Lx*1e-9)*(Ly*1e-9)*(Lz*1e-9)
    hh *= (cte.e/m)**2

    return hh * rho
```

```python
from matplotlib.collections import PolyCollection
from matplotlib import colors as mcolors

fig1 = plt.figure(dpi=200)
fig2 = plt.figure(dpi=200)
ax = fig2.gca(projection='3d')
ax.xaxis.set_tick_params(labelsize=5)
ax.yaxis.set_tick_params(labelsize=5)
ax.zaxis.set_tick_params(labelsize=5)


kSubPlot = 0
dkS = numSimulations/4.0
nkS = 0
absorption_vs_energy = []
a_max = 0

for k in range(numSimulations): # nm
    Lx = Lx_values[k]
    a = absorption(Lx)/100 # cm^-1
    a_max = np.max([a_max, a.max()])
    absorption_vs_energy.append(list(zip(E_photon, a)))
    ax.plot(E_photon, a, 'k', zs=Lx, zdir='y')
    ax.set_xlabel(r'$\hbar\omega$    $[eV]$')
    ax.set_ylabel(r'$L_x$  $[nm]$')
    ax.set_zlabel(r'$\alpha(\omega)$    $[cm^{-1}]$')
    ax.view_init(elev=50., azim=85)

    if k >= nkS:
        nkS += dkS
        kSubPlot += 1
        plt.figure(fig1.number)
        plt.subplot(2,2,kSubPlot)
        plt.plot(E_photon, a, 'k', lw=1)
        plt.xlabel(r'$\hbar\omega$    $[eV]$')
        plt.ylabel(r'$\alpha(\omega)$    $[cm^{-1}]$')
        #plt.grid(which='both', axis='both', color='g')
        plt.title(r'$L_x = %0.1f nm$' % Lx)

plt.figure(fig1.number)
plt.subplots_adjust(hspace=0.8, wspace=0.5)
plt.show()

plt.figure(fig2.number)
plt.show()
```

```python
fig = plt.figure(dpi=200)
ax  = fig.gca(projection='3d')
ax.xaxis.set_tick_params(labelsize=5)
ax.yaxis.set_tick_params(labelsize=5)
ax.zaxis.set_tick_params(labelsize=5)

allAbsorptions = PolyCollection(absorption_vs_energy)
allAbsorptions.set_alpha(0.7)
ax.add_collection3d(allAbsorptions, zs=Lx_values, zdir='y')
ax.set_xlabel(r'$\hbar\omega$    $[eV]$')
ax.set_xlim3d(E_min, E_max)
ax.set_ylabel(r'$L_x$  $[nm]$')
ax.set_ylim3d(Lx_min, Lx_max)
ax.set_zlabel(r'$\alpha(\omega)$    $[cm^{-1}]$')
ax.set_zlim3d(0, a_max)
ax.view_init(elev=50., azim=85)
```
